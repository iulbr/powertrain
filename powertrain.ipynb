# Powertrain
Powertrain implementation

[Folge natürlicher Zahlen bei OEIS](https://oeis.org/A133500)

[Freistätters Formelwelt: **Warum Zahlen seltsam sind**](https://www.spektrum.de/kolumne/warum-zahlen-seltsam-sind/1660094)

```python
def powertraining(von,bis):
    for x in range(von,bis+1):
        zähler = 0;
        # Zähler, damit der Rechner nicht abstürzt bei
        # z.B 24547284284866560000000000 und 2592
        print("\nPowertraining\n" + str(x))
        while x > 9 and zähler < 10:
        # Solange x nicht einstellig wiederhole Folgendes: 
            x = [int(j) for j in str(x)] #x in Ziffernliste wandeln
            x_ = 1 # Ausgabe, x_ bedeutet hier x', auf 1 setzen
            for i in range(0, len(x)-1, 2):
            # Im 2er-Schritt durch die Ziffernliste
                x_ *= x[i] ** x[i + 1]
                # bei jedem Schritt Ziffer mit Nachfolger potenzieren und Gesamtprodukt bilden
            if len(x) % 2 == 1:
                x_ *= x[len(x)-1] 
            zähler += 1
            print(x_)
            x=x_

powertraining(8,22)
```
